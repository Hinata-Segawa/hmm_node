```python
#(S,F)としたときにS+F = xとする． S = 弦，F = フレット ，1から126まで

observe_states, frets = guitar_frets_dict()
n_observe_states = len(observe_states)

# print(frets[1].keys())  #キーのみ出力
# print(frets[1].values())  #値のみの出力
# E4 = [d.get('E4') for d in frets]#E4の数字取得
# print(E4)

```

```python
observes = [[23,47,22,30]] #23,47,22,30の順番の観測
n_observes = len(observes)

print(n_observe_states)
```

```python
# 初期状態確率，全部同じ
startprob = []
for i in range(n_observe_states):
    startprob.append(1/n_observe_states)

startprob = np.array(startprob)
startprob
```

```python
frets_items = []

#全部のキーと要素を抽出
for i in range(6):
  frets_items.append(list(frets[i].items()))

print(frets_items)

#二次元配列になっているので平坦化する
frets_items = list(itertools.chain.from_iterable(frets_items))

#番号順にソートする
frets_items = sorted(frets_items, key = lambda x:x[1])
print(frets_items)

frets_items[3][0]
# name_list[1]
```

```python
# 出力確率
emmisionprob = []

sum_list = [0]*n_observe_states #各行の数値の合計値

#100,80,1の割り当て
for i in range(n_observe_states):
  i_key = frets_items[i][0]
  for j in range(n_observe_states):
    if i == j:#自分のキーと要素が同じなら100
      emmisionprob.append(100)
      sum_list[i] += 100

    elif frets_items[j][0] == i_key:#同じ音なら80
      emmisionprob.append(80)
      sum_list[i] += 80

    else:
      emmisionprob.append(1)
      sum_list[i] += 1

#126*126の行列にする
emmisionprob = np.array(emmisionprob).reshape(n_observe_states,n_observe_states)

#numpy配列はint型だからfloatにする
emmisionprob = emmisionprob.astype(np.float64)



#正規化
for i in range(n_observe_states):
  for j in range(n_observe_states):
    emmisionprob[i][j] = emmisionprob[i][j]/sum_list[i]

print(emmisionprob[0])

# emmisionprob[0][0]

#観測変数と潜在変数が同じ時だけ1にする．
# for i in range(n_observe_states):
#   for j in range(n_observe_states):
#       if j == i:
#         emmisionprob.append(1.0-(0.001*(n_observe_states-1)))
#       else:
#         emmisionprob.append(0.001)

# emmisionprob = np.array(emmisionprob).reshape(n_observe_states,n_observe_states)

# emmisionprob

```

```python
# 遷移確率
transmat = []

for i in range(n_observe_states):
  for j in range(n_observe_states):
    transmat.append(1/n_observe_states)

transmat = np.array(transmat).reshape(n_observe_states,n_observe_states)
transmat
```

```python
# hmmlearn のインスタンス作成
model = hmm.MultinomialHMM(n_components = n_observe_states)

model.n_features = n_observe_states #状態の数
model.startprob_= startprob #初期確率
model.transmat_= transmat #遷移確率
model.emissionprob_= emmisionprob #出力確率

```

```python
order_node = np.array(observes).T


# 推定
logprob, decoded_codes = model.decode(order_node)
print(decoded_codes)

# print("観測:")
# for i in range(len(order_node)):
#   print()

```
